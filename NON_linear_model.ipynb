```python
import tensorflow
from tensorflow.keras.models import Model
```

```python
from keras.layers import *
```

```python
InputLayer=Input(shape=(3,))
```

```python
DenseLayer1=Dense(32,activation='relu')(InputLayer)
DenseLayer2=Dense(64,activation='relu')(DenseLayer1)
DenseLayer3=Dense(96,activation='relu')(DenseLayer2)
sublayer1=Dense(32,activation='relu')(DenseLayer3)
sublayer2=Dense(64,activation='relu')(DenseLayer3)
output1=Dense(1,activation='linear')(sublayer1)
output2=Dense(1,activation='sigmoid')(sublayer2)
```

```python
model=Model(inputs=InputLayer , outputs =[output1,output2] )
```

```python
from keras.utils import plot_model
```

```python
plot_model(model=model,show_dtype=True,show_shapes=True)
```
